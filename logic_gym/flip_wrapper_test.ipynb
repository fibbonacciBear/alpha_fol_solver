```python
import sys

PACKAGE_HOME = "../../alpha_fol_solver"
sys.path.append(f"{PACKAGE_HOME}")

from gym_natural_deduction import FlipWrapper
```

```python
# This problem will produce true as the final goal_state

axioms = """
Max = Variable("Max")


class yumpus(Relation):
    def __init__(self, *args):  # seq of args
        Relation.__init__(self, *args)
        

class dumpus(Relation):
    def __init__(self, *args):  # seq of args
        Relation.__init__(self, *args)
        

class wumpus(Relation):
    def __init__(self, *args):  # seq of args
        Relation.__init__(self, *args)
        

class impus(Relation):
    def __init__(self, *args):  # seq of args
        Relation.__init__(self, *args)
        

class timpus(Relation):
    def __init__(self, *args):  # seq of args
        Relation.__init__(self, *args)
        

class sour(Relation):
    def __init__(self, *args):  # seq of args
        Relation.__init__(self, *args)
        
checkp(Text("test2"), comment)
checkp(A(x, Impl(yumpus(x), (dumpus(x)))), given)
checkp(A(x, Impl(dumpus(x), (wumpus(x)))), given)
checkp(A(x, Impl(wumpus(x), (impus(x)))), given)
checkp(A(x, Impl(impus(x), (timpus(x)))), given)
checkp(A(x, Impl(timpus(x), Not(sour(x)))), given)
checkp(yumpus(Max), given)
checkp(Not(sour(Max)), goal)

"""

axioms_list = axioms.strip().split("\n")
axioms_list = [elem if not elem.isspace() else "" for elem in axioms_list]


proof = """
rapply(Ae, 1, Max)
rapply(imple, 8, 6)
rapply(Ae, 2, Max)
rapply(imple, 10, 9)
rapply(Ae, 3, Max)
rapply(imple, 12, 11)
rapply(Ae, 4, Max)
rapply(imple, 14, 13)
rapply(Ae, 5, Max)
rapply(imple, 16, 15)

"""

proof_steps = proof.strip().split("\n")
```

```python
# Run the proof programmatically through the Flip wrapper

wrap = FlipWrapper()

# start

try:
    state = wrap.start(axioms_list)
    print(state)
except Exception as e:
    print(f"An error occurred: {str(e)}")
    
print("Ready to check a proof...")

try:
    for proof_step in proof_steps:
        goal_state, state = wrap.run_proof_step(proof_step)
        print("State:" )
        print(state)
        print("goal_state: " + goal_state + "\n")
except Exception as e:
    print(f"An error occurred: {str(e)}")

wrap.terminate()

assert goal_state == "true", "You broke something..."
```
